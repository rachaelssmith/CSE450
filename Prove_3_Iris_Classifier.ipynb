```
#classification of Fisher's Iris dataset
#import the following files
import random
import pandas as pd
import numpy as np
import collections 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
#load the data

url = "https://byui-cs.github.io/cs450-course/week01/iris.data"
df = pd.read_csv(url, header=0)
print(df)
c = collections.Counter()
```

```
#randomize data
df = df.sample(frac=1)
df
```

```
#split the data into features and targets
targets = df['species']
features = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
print(targets)
print(features)

targets = targets.to_numpy()
features =features.to_numpy()
print("Targets to Numpy: ", targets)
print("Features to Numpy: ", features)


#randomize and split the samples into two groups.
#30% of the samples will be used for testing.
#the other 70% will be used for training.
train_data, test_data, train_targets, test_targets = train_test_split(features, targets, test_size =.3)

print ("train data shape: "+str(train_data.shape+train_targets.shape))
print ("test data shape: "+str(test_data.shape+test_targets.shape))
```

```
class HardCodedClassifier:
    def __init__(self, k):
        self.tr_data = None
        self.tr_targets = None
        self.ts_data = None
        self.ts_targets = None
        self.k = k
        self.x = None
        self.y = None
        self.predictions = []
        
    def fit(self, tr_data, tr_targets):
        """
        Receives: training data and training targets
        Returns: None
        """
        self.tr_data = tr_data
        self.tr_targets = tr_targets

    def predict(self, ts_data):
        """
        Receives: test data (a bunch of test rows)
        Returns: list of predictions for the targets associated with each row
        in the test data
        returns test_targets
        """
        
        self.ts_data = ts_data
        for row in ts_data:
            print(row)
            self.ts_prediction = self.predict_single(row)
            print("Single row prediciton: ", self.ts_prediction)
            lst = self.ts_targets.np.append(self.ts_targets)

    def predict_single(self, test_row):
        """
        Gets: a single row
        Returns: single row of data for that row
        """
        # find k rows that are the closest
        # we want the target for k not distance
        # choose most common target in list of top K
        c = collections.Counter()
        predict_single_list = []
        k_possible_predictions = []
        self.y = test_row
        for row in self.tr_data:
          predict_single_list.append(self.calc_distance(row, test_row))
          sorted_predict_single_index = np.argsort(predict_single_list)
      
        k_possible_predictions = sorted_predict_single_index[:self.k]
        np.array(k_possible_predictions)

        self.ts_targets = collections.Counter(k_possible_predictions).most_common(1)
        print(self.ts_targets)
        
        #print("sorted predict single index range K: ", sorted_predict_single_index[:k])
        
        #print("Printing K:", sorted_predict_single_index[0:k])
        for i in k_possible_predictions:
          while self.k != 0:
            self.ts_targets.append(self.tr_targets[0])
            self.k -= 1
        self.ts_targets = collections.Counter(self.ts_targets).most_common(1)[0][0]
        print(self.ts_targets)
        
        #print("test targets: \n", self.ts_targets[:])
          
    def calc_distance(self, x, y):
        self.x = x
        self.y = y
        return ((x[0] - y[0]) ** 2 + (x[1] - y[1]) ** 2 + (x[2] - y[2]) ** 2 + (x[3] - y[3]) ** 2)

```

```
def main():
    # trying to see results
    k = 3
    hcc = HardCodedClassifier(k)
    hcc.fit(train_data, train_targets)
    iris_hcc_predictions = hcc.predict(test_data)
    print(iris_hcc_predictions)

if __name__ == "__main__":
  main()
```
