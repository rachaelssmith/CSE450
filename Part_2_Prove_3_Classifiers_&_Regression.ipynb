```
#classification of Math dataset

#import the following files
import os
import random
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score

#see full columns in pandas.

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
#load the data
```

```
#Import the Math dataset

math_names = ["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","guardian","traveltime","studytime","failures","schoolsup","famsup","paid","activities","nursery","higher","internet","romantic","famrel","freetime","goout","Dalc","Walc","health","absences","G1","G2","G3"]
url = ('https://docs.google.com/spreadsheets/d/e/2PACX-1vRgvwz3Do6-7SWmN9PwYmyopu7w7QSYF1ANGz1zEudToL8Y6SCedrhfUY91Qm-zp1EEY8SBdttQX9pa/pub?output=csv')
mathdf = pd.read_csv(url, dtype = str, index_col = False, skiprows = 1, delimiter = ",", names = math_names)

mathdf
```

```
# Cleanup the Math dataset

# Changing objects to numerical values
# Cleanup the binary objects into categories:
mathdf["school"] = mathdf["school"].astype('category')
mathdf["school"] = mathdf["school"].cat.codes
mathdf["sex"] = mathdf["sex"].astype('category')
mathdf["sex"] = mathdf["sex"].cat.codes
mathdf["address"] = mathdf["address"].astype('category')
mathdf["address"] = mathdf["address"].cat.codes
mathdf["famsize"] = mathdf["famsize"].astype('category')
mathdf["famsize"] = mathdf["famsize"].cat.codes
mathdf["Pstatus"] = mathdf["Pstatus"].astype('category')
mathdf["Pstatus"] = mathdf["Pstatus"].cat.codes
mathdf["schoolsup"] = mathdf["schoolsup"].astype('category')
mathdf["schoolsup"] = mathdf["schoolsup"].cat.codes
mathdf["famsup"] = mathdf["famsup"].astype('category')
mathdf["famsup"] = mathdf["famsup"].cat.codes
mathdf["paid"] = mathdf["paid"].astype('category')
mathdf["paid"] = mathdf["paid"].cat.codes
mathdf["activities"] = mathdf["activities"].astype('category')
mathdf["activities"] = mathdf["activities"].cat.codes
mathdf["nursery"] = mathdf["nursery"].astype('category')
mathdf["nursery"] = mathdf["nursery"].cat.codes
mathdf["higher"] = mathdf["higher"].astype('category')
mathdf["higher"] = mathdf["higher"].cat.codes
mathdf["internet"] = mathdf["internet"].astype('category')
mathdf["internet"] = mathdf["internet"].cat.codes
mathdf["romantic"] = mathdf["romantic"].astype('category')
mathdf["romantic"] = mathdf["romantic"].cat.codes
mathdf["Mjob"] = mathdf["Mjob"].astype('category')
mathdf["Mjob"] = mathdf["Mjob"].cat.codes
mathdf["Fjob"] = mathdf["Fjob"].astype('category')
mathdf["Fjob"] = mathdf["Fjob"].cat.codes
mathdf["reason"] = mathdf["reason"].astype('category')
mathdf["reason"] = mathdf["reason"].cat.codes
mathdf["guardian"] = mathdf["guardian"].astype('category')
mathdf["guardian"] = mathdf["guardian"].cat.codes

mathdf["studytime"] = pd.to_numeric(mathdf["studytime"])
mathdf["failures"] = pd.to_numeric(mathdf["failures"])
mathdf["Medu"] = pd.to_numeric(mathdf["Medu"])
mathdf["Fedu"] = pd.to_numeric(mathdf["Fedu"])
mathdf["age"] = pd.to_numeric(mathdf["age"])
mathdf["traveltime"] = pd.to_numeric(mathdf["traveltime"])
mathdf["famrel"] = pd.to_numeric(mathdf["famrel"])
mathdf["freetime"] = pd.to_numeric(mathdf["freetime"])
mathdf["goout"] = pd.to_numeric(mathdf["goout"])
mathdf["Dalc"] = pd.to_numeric(mathdf["Dalc"])
mathdf["Walc"] = pd.to_numeric(mathdf["Walc"])
mathdf["health"] = pd.to_numeric(mathdf["health"])
mathdf["absences"] = pd.to_numeric(mathdf["absences"])
mathdf["G1"] = pd.to_numeric(mathdf["G1"])
mathdf["G2"] = pd.to_numeric(mathdf["G2"])
mathdf["G3"] = pd.to_numeric(mathdf["G3"])

mathdf
```

```
# Randomize the Math Data
mathdf = mathdf.sample(frac=1)

#split the data into features and targets
math_targets = mathdf['G3']
math_features = mathdf[math_names[:32]]

print(math_features)

math_targets = math_targets.to_numpy()
math_features = math_features.to_numpy()

#30% of the samples will be used for testing.
#the other 70% will be used for training.
math_train_data, math_test_data, math_train_targets, math_test_targets = train_test_split(math_features, math_targets, test_size =.3)
```

```
def main():
  knnregression = KNeighborsRegressor(n_neighbors=3)
  knnregression.fit(math_train_data, math_train_targets)
  math_predictions = knnregression.predict(math_test_data)
  y_true2 = math_test_targets
  y_pred2 = math_predictions
  math_score = r2_score(y_true2, y_pred2)

  print("\nMath Test Targets: \n", math_test_targets, "\n\nMath Regression Predictions: \n", math_predictions)
  print ("Math train data shape: " + str(math_train_data.shape+math_train_targets.shape))
  print ("\nMath test data shape: " + str(math_test_data.shape+math_test_targets.shape)) 
  print("\nMath Regression Score: {:.2f}".format(math_score))
  
if __name__ == "__main__":
    main()
```
